```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(4,3))
sns.lineplot(data=df, x='studytime', y='failures')
plt.title("Graph based on Study time vs Failures")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.lineplot(data=df, x='absences', y='failures')
plt.title("Graph based on Absence vs Failures")
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(df_correlation_data, annot=True, cmap='Purples', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(df_data, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(4,3))
p=df['school'].value_counts()
plt.pie(p,labels=p.index,autopct="%0.2f%%", colors=['Blue', 'orange'])
plt.title("School Distribution")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='absences',data=df, color='green')
plt.title("Distribution of Absences")
plt.show()
```
